```python
import pandas as pd
cpi = pd.read_csv('cpi.csv')
dxy = pd.read_csv('DXY.csv')
gdp = pd.read_csv('GDP.csv')
fedfundrate = pd.read_csv('FedFundRate.csv')
nasdaq = pd.read_csv('NASDAQ.csv')

# Convert the date columns to datetime format
cpi['DATE'] = pd.to_datetime(cpi['DATE'])
dxy['Date'] = pd.to_datetime(dxy['Date'])
gdp['DATE'] = pd.to_datetime(gdp['DATE'])
fedfundrate['DATE'] = pd.to_datetime(fedfundrate['DATE'])
nasdaq['DATE'] = pd.to_datetime(nasdaq['DATE'])
cpi.set_index('DATE', inplace=True)
dxy.set_index('Date', inplace=True)
gdp.set_index('DATE', inplace=True)
fedfundrate.set_index('DATE', inplace=True)
nasdaq.set_index('DATE', inplace=True)
```

```python
#Monthly frequent
cpi = cpi.resample('3M').ffill()
dxy = dxy.resample('3M').ffill()
gdp = gdp.resample('3M').ffill()
fedfundrate = fedfundrate.resample('3M').ffill()
nasdaq = nasdaq.resample('3M').ffill()
```

```python
cpi_normalized = cpi.diff(periods=1)/cpi
dxy_normalized_one_month = dxy.diff(periods=1)/dxy
gdp_normalized_one_month = gdp.diff(periods=1)/gdp
```

```python
cpi_normalized_one_month=cpi_normalized_one_month.dropna()
dxy_normalized_one_month=dxy_normalized_one_month.dropna()
gdp_normalized_one_monthh=gdp_normalized_one_month.dropna()
# cpi_normalized_two_month = cpi_normalized_two_month.dropna()
# dxy_normalized_two_month = dxy_normalized_two_month.dropna()
# gdp_normalized_two_month = gdp_normalized_two_month.dropna()
# cpi_normalized_three_month = cpi_normalized_three_month.dropna()
# dxy_normalized_three_month = dxy_normalized_three_month.dropna()
# gdp_normalized_three_month = gdp_normalized_three_month.dropna()
```

```python
system_state = pd.DataFrame()

system_state['CPI'] = cpi_normalized.cumsum()
system_state['CPI_90_mean'] = cpi_normalized.rolling(window='90D').mean()
system_state['CPI_90_max'] = cpi_normalized.rolling(window='90D').max()
system_state['CPI_90_min'] = cpi_normalized.rolling(window='90D').min()
system_state['CPI_180_mean'] = cpi_normalized.rolling(window='180D').mean()
system_state['CPI_180_max'] = cpi_normalized.rolling(window='180D').max()
system_state['CPI_180_min'] = cpi_normalized.rolling(window='180D').min()
system_state['CPI_270_mean'] = cpi_normalized.rolling(window='270D').mean()
system_state['CPI_270_max'] = cpi_normalized.rolling(window='270D').max()
system_state['CPI_270_min'] = cpi_normalized.rolling(window='270D').min()
system_state['CPI_365_mean'] = cpi_normalized.rolling(window='365D').mean()
system_state['CPI_365_max'] = cpi_normalized.rolling(window='365D').max()
system_state['CPI_365_min'] = cpi_normalized.rolling(window='365D').min()
system_state['DXY'] = dxy_normalized_one_month.cumsum()
system_state['DXY_90_mean'] = dxy_normalized_one_month.rolling(window='90D').mean()
system_state['DXY_90_max'] = dxy_normalized_one_month.rolling(window='90D').max()
system_state['DXY_90_min'] = dxy_normalized_one_month.rolling(window='90D').min()
system_state['DXY_180_mean'] = dxy_normalized_one_month.rolling(window='180D').mean()
system_state['DXY_180_max'] = dxy_normalized_one_month.rolling(window='180D').max()
system_state['DXY_180_min'] = dxy_normalized_one_month.rolling(window='180D').min()
system_state['DXY_270_mean'] = dxy_normalized_one_month.rolling(window='270D').mean()
system_state['DXY_270_max'] = dxy_normalized_one_month.rolling(window='270D').max()
system_state['DXY_270_min'] = dxy_normalized_one_month.rolling(window='270D').min()
system_state['DXY_365_mean'] = dxy_normalized_one_month.rolling(window='365D').mean()
system_state['DXY_365_max'] = dxy_normalized_one_month.rolling(window='365D').max()
system_state['DXY_365_min'] = dxy_normalized_one_month.rolling(window='365D').min()
system_state['FEDFUNDRATE']  = fedfundrate.cumsum()
system_state['FEDFUNDRATE_90_mean']  = fedfundrate.rolling(window='90D').mean()
system_state['FEDFUNDRATE_90_max']  = fedfundrate.rolling(window='90D').max()
system_state['FEDFUNDRATE_90_min']  = fedfundrate.rolling(window='90D').min()
system_state['FEDFUNDRATE_180_mean']  = fedfundrate.rolling(window='180D').mean()
system_state['FEDFUNDRATE_180_max']  = fedfundrate.rolling(window='180D').max()
system_state['FEDFUNDRATE_180_min']  = fedfundrate.rolling(window='180D').min()
system_state['FEDFUNDRATE_270_mean']  = fedfundrate.rolling(window='270D').mean()
system_state['FEDFUNDRATE_270_min']  = fedfundrate.rolling(window='270D').min()
system_state['FEDFUNDRATE_270_max']  = fedfundrate.rolling(window='270D').max()
system_state['FEDFUNDRATE_365_mean']  = fedfundrate.rolling(window='365D').mean()
system_state['FEDFUNDRATE_365_max']  = fedfundrate.rolling(window='365D').max()
system_state['FEDFUNDRATE_365_min']  = fedfundrate.rolling(window='365D').min()
system_state['GDP']  = gdp
```

# using wcss to check clustring is good or not you can see refrence in https://analyticsindiamag.com/beginners-guide-to-k-means-clustering/

```python
from sklearn.cluster import KMeans
import seaborn as sns
wcss = {}
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300)
    kmeans.fit(X)
    wcss[k] = kmeans.inertia_

sns.pointplot(x=list(wcss.keys()), y=list(wcss.values()))
```

همانطور که مشخص است بهتر است تعداد خوشه های 5 یا 6 در نظر بگیریم زیرا از آنجا به بعد تغییر خیلی زیادی نداریم . در این اینجا تعداد خوشه ها را برابر 6 در نظر میگیریم

```python

kmeans = KMeans(n_clusters=6, init='k-means++', max_iter=300)
kmeans.fit(X)
```

```python
data['cluster'] = kmeans.labels_
```

```python
data['GDP']=gdp_normalized_one_month
```

```python
gdp_stats = data.groupby('cluster')['GDP'].agg(['median', 'var'])
```

```python
system_state = pd.DataFrame()
system_state['CPI_365'] = cpi_normalized.rolling(window='365D').mean()
system_state['FEDFUNDRATE_365']  = fedfundrate.rolling(window='365D').mean()
system_state['DXY_365'] = dxy_normalized_one_month.rolling(window='365D').mean()
```

```python
system_state['GDP']  = gdp
```

```python
data = system_state.dropna()
X = data.drop(['GDP'], axis=1)
from sklearn.cluster import KMeans
import seaborn as sns
wcss = {}
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=300)
    kmeans.fit(X)
    wcss[k] = kmeans.inertia_

sns.pointplot(x=list(wcss.keys()), y=list(wcss.values()))
```

```python
kmeans = KMeans(n_clusters=6, init='k-means++', max_iter=300)
kmeans.fit(X)
data['cluster'] = kmeans.labels_
data["GDP"] = gdp_normalized_one_month
gdp_stats = data.groupby('cluster')['GDP'].agg(['median', 'var'])
```
